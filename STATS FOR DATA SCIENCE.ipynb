```python
## QUESTION 1:  Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)
```

```python
df = df.replace('-', df.replace(['-'], [0]))
```

```python
df.info()
```

```python
## Converting objects to string
```

```python
df["Points"] = df["Points"].astype(int)
df["Seasons"] = df["Seasons"].astype(int)
df["Points"] = df["Points"].astype(int)
df["GamesPlayed"] = df["GamesPlayed"].astype(int)
df["GamesWon"] = df["GamesWon"].astype(int)
df["GamesDrawn"] = df["GamesDrawn"].astype(int)
df["GamesLost"] = df["GamesLost"].astype(int)
df["GoalsFor"] = df["GoalsFor"].astype(int)
df["GoalsAgainst"] = df["GoalsAgainst"].astype(int)
df["Champion"] = df["Champion"].astype(int)
df["Runner-up"] = df["Runner-up"].astype(int)
df["Third"] = df["Third"].astype(int)
df["Fourth"] = df["Fourth"].astype(int)
df["Fifth"] = df["Fifth"].astype(int)
df["Sixth"] = df["Sixth"].astype(int)
df["T"] = df["T"].astype(int)
df["BestPosition"] = df["BestPosition"].astype(int)   
```

```python
df.info()
```

```python
df.head()
```

```python
## Replacing the Debut & Since/Last appearance with valid details
df['Debut'] = df['Debut'].apply(lambda x: x[:4] if ("-" in x) else x)
```

```python
df['Since/LastApp'] = df['Since/LastApp'].apply(lambda x: (x[:2] + x[-2:]) if ("-" in x) else x)
```

```python
# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#  Using the same function, find the team which has maximum and minimum goal difference. (5 points)
```

```python
def Goal_diff_count():
    # Goal_diff_count = GoalsFor - GoalsAgainst
    df['Goal_diff_count'] = df['GoalsFor'] - df['GoalsAgainst']
```

```python
Goal_diff_count()
```

```python
# Maximum goal difference
df[df['Goal_diff_count'] == df['Goal_diff_count'].max()]
```

```python
# Minimum goal difference
df[df['Goal_diff_count'] == df['Goal_diff_count'].min()]
```

```python
# 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)

# Percentage of Winning = (GamesWon / GamesPlayed)*100

# If there are any numerical error, replace it with 0%

# Print the top 5 teams which has the highest Winning percentage
```

```python
df['Winning Percent'] = (df['GamesWon'] / df['GamesPlayed']) * 100
```
